```
df = df[['TOP_HOT_AIR_TEMP', 'CHAMBER_EXT_TEMP','CHAMBER_OUTLET_TEMP', 'FEED_PUMP_FLOW']]
```

```
df_final = pd.DataFrame(columns = df.columns)
```

```
ff = dff.sample(n = 2500, replace = True)
```

```
faulty_top = ff[['Faulty_TOP_HOT_AIR_TEMP', 'CHAMBER_EXT_TEMP', 'CHAMBER_OUTLET_TEMP', 'FEED_PUMP_FLOW']]
```

```
faulty_exit = ff[['TOP_HOT_AIR_TEMP', 'Faulty_CHAMBER_EXT_TEMP', 'CHAMBER_OUTLET_TEMP', 'FEED_PUMP_FLOW']]
```

```
faulty_outlet = ff[['TOP_HOT_AIR_TEMP', 'CHAMBER_EXT_TEMP', 'Faulty_CHAMBER_OUTLET_TEMP', 'FEED_PUMP_FLOW']]
```

```
faulty_feed = ff[['TOP_HOT_AIR_TEMP', 'CHAMBER_EXT_TEMP', 'CHAMBER_OUTLET_TEMP', 'Faulty_FEED_PUMP_FLOW']]
```

```
faulty_top.rename(columns = {'Faulty_TOP_HOT_AIR_TEMP': 'TOP_HOT_AIR_TEMP'}, inplace = True)
```

```
faulty_exit.rename(columns = {'Faulty_CHAMBER_EXT_TEMP': 'CHAMBER_EXT_TEMP'}, inplace = True)
faulty_outlet.rename(columns = {'Faulty_CHAMBER_OUTLET_TEMP': 'CHAMBER_OUTLET_TEMP'}, inplace = True)
faulty_feed.rename(columns = {'Faulty_FEED_PUMP_FLOW': 'FEED_PUMP_FLOW'}, inplace = True)
```

```
faulty_top['target'] = np.ones(len(faulty_top.TOP_HOT_AIR_TEMP)).astype(int)
```

```
faulty_top
```

```
faulty_feed['target'] = 4*np.ones(len(faulty_top.TOP_HOT_AIR_TEMP)).astype(int)
```

```
faulty_outlet['target'] = 3*np.ones(len(faulty_top.TOP_HOT_AIR_TEMP)).astype(int)
```

```
faulty_exit['target'] = 2*np.ones(len(faulty_top.TOP_HOT_AIR_TEMP)).astype(int)
```

```
df_final = df_final.append(faulty_top)
df_final = df_final.append(faulty_exit)
df_final = df_final.append(faulty_outlet)
df_final = df_final.append(faulty_feed)
```

```
df_final = df_final.append(df)
```

```
df_final['target'] = df_final['target'].fillna(int(0))
```

```
df_final.target = df_final.target.astype(int)
```
